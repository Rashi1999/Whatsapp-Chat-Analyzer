```python
import re
import pandas as pd
```

```python
f=open('WhatsApp Chat with Eshita Christ BCA.txt','r',encoding='utf-8')
```

```python
data=f.read()
```

```python
print(data)
```

```python
print(type(data))
```

```python
pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
#to remove the first empty string in the list we took [1:]
```

```python
messages
```

```python
dates = re.findall(pattern, data)
```

```python
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
```

```python
df
# date in the format string- 'mm/dd/yy, hh:mm - '
```

```python
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %H:%M - ')
```

```python
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
df.head(20)
#date in the format date - yyyy-mm-dd hh:mm:ss
```

```python
df.shape
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df.head()
```

```python
# dt can be used to access the values of the series as datetimelike 
# and return several properties. Pandas Series. dt. minute attribute 
# return a numpy array containing the minutes of the datetime in the 
# underlying data of the given series object.

# df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
# df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
# df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
df.head()
```
